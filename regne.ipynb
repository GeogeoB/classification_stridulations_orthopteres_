```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import community 
```

```python
nom_fichier = './data/infos/TAXREF16.0_FR__14_11_2023.csv'

# Charger le fichier CSV dans un DataFrame Pandas
data = pd.read_csv(nom_fichier, encoding='latin-1', sep=';')

# Afficher les premières lignes du DataFrame pour vérification
print(data.head())
```

```python
ORDRE = data.groupby('ORDRE')
print(f'Il y a {len(ORDRE.size())} familles différentes \n')
print(ORDRE.size())
```

```python
FAMILLE = data.groupby('FAMILLE')
nb_element_FAMILLE = FAMILLE.size()
print(f'Il y a {len(FAMILLE.size())} familles différentes \n')
print(nb_element_FAMILLE)
```

```python
SOUS_FAMILLE = data.groupby('SOUS_FAMILLE')
nb_element_SOUS_FAMILLE = SOUS_FAMILLE.size()

print(f'Il y a {len(SOUS_FAMILLE.size())} sous-familles différentes \n')
print(nb_element_SOUS_FAMILLE)
```

```python
TRIBU = data.groupby('TRIBU')
nb_element_TRIBU = TRIBU.size()

print(f'Il y a {len(TRIBU.size())} tribus différentes \n')
print(nb_element_TRIBU)
```

```python
def afficher_pandas(element):
    # Création de l'histogramme à partir des données
    plt.figure(figsize=(10, 6))  # Taille de la figure
    element.plot(kind='bar', color='skyblue')  # Création du graphique en barres
    plt.xlabel('Sous-famille')  # Nom de l'axe des x
    plt.ylabel('Nombre d\'éléments')  # Nom de l'axe des y
    plt.title('Histogramme du nombre d\'éléments par sous-famille')  # Titre du graphique
    plt.xticks(rotation=90)  # Rotation des étiquettes sur l'axe x pour une meilleure lisibilité
    plt.tight_layout()  # Ajustement de la mise en page
    plt.show()  # Affichage du graphique
```

```python
afficher_pandas(nb_element_FAMILLE)
afficher_pandas(nb_element_SOUS_FAMILLE)
afficher_pandas(nb_element_TRIBU)
```

```python
G = nx.DiGraph()

# Ajout des arêtes (relations) entre les différentes catégories taxonomiques
# (Supposons que data est le DataFrame contenant les relations entre les catégories taxonomiques)
for _, row in data.iterrows():
    G.add_edge(row['REGNE'], row['PHYLUM'])
    G.add_edge(row['PHYLUM'], row['CLASSE'])
    G.add_edge(row['CLASSE'], row['ORDRE'])
    G.add_edge(row['ORDRE'], row['FAMILLE'])
    G.add_edge(row['FAMILLE'], row['SOUS_FAMILLE'])
    G.add_edge(row['SOUS_FAMILLE'], row['TRIBU'])

for node in G.nodes():
    if node in data['REGNE'].values:
        G.nodes[node]['subset'] = 0
    elif node in data['PHYLUM'].values:
        G.nodes[node]['subset'] = 1
    elif node in data['CLASSE'].values:
        G.nodes[node]['subset'] = 2
    elif node in data['ORDRE'].values:
        G.nodes[node]['subset'] = 3
    elif node in data['FAMILLE'].values:
        G.nodes[node]['subset'] = 4
    elif node in data['SOUS_FAMILLE'].values:
        G.nodes[node]['subset'] = 5
    elif node in data['TRIBU'].values:
        G.nodes[node]['subset'] = 6
    else :
        G.nodes[node]['subset'] = 7

# Dessiner le graphe
pos = nx.multipartite_layout(G)
plt.figure(figsize=(12, 10))
nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_weight='bold', font_size=5)
plt.title('Hiérarchie Tribu par rapport aux Sous-familles')
plt.show()
```
